```python
import pandas as pd
import os
import numpy as np
```

```python
WINEPATH = './data/'
```

```python
def load_wine_data(wine_path = WINEPATH):
    csv_path = os.path.join(wine_path, 'winequality-red.csv')
    return pd.read_csv(csv_path)
```

```python
wines = load_wine_data()

wines.head()

wines.info()

wines.describe()

%matplotlib inline
import matplotlib.pyplot as plt
wines.hist(bins=50, figsize=(20,15))
plt.show()

wines.shape

data_points = len(wines)* 12
print(data_points)
```

```python
def split_train_test(data, test_ratio):
    np.random.seed(42) # create same test set each time
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
# more compact train test split from sklearn

from sklearn.model_selection import train_test_split

train_set , test_set = train_test_split(wines, test_size = 0.2, random_state = 42)
```
